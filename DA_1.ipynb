```python
#Loading Corpus using file method
corpus = []
filename = os.listdir('/content/drive/MyDrive/Dataset')
for i in filename:
  file = open('/content/drive/MyDrive/Dataset/'+i,'r')
  text = file.read()
  corpus.append(text)
  file.close()
print(corpus)
```

```python
#Loading corpus using plaintextCorpusReader
from nltk.corpus import PlaintextCorpusReader
cor = PlaintextCorpusReader('/content/drive/MyDrive/Dataset/','.*')
cor.words()
```

```python
#Normalization
lower = []
for i in range(len(corpus)):
  lower.append(" ".join(word.lower() for word in corpus[i].split()))
print(lower)
alpha = []
for i in range(len(lower)):
  alpha.append(" ".join(word for word in lower[i].split() if word.isalpha()))
print(alpha)
```

```python
#Tokenization
from nltk import word_tokenize
tokenized = []
for i in range(len(alpha)):
  tokenized.append(word_tokenize(alpha[i]))
print(tokenized)
```

```python
#StopWord removal
from nltk.corpus import stopwords
stop = stopwords.words(fileids='english')
no_stop = []
for i in range(len(tokenized)):
  no_stop.append([word for word in tokenized[i] if word not in stop])
print(no_stop)
```

```python
#Stemming
from nltk.stem import PorterStemmer
no_stem = []
ps = PorterStemmer()
for i in range(len(no_stop)):
  no_stem.append(' '.join([ps.stem(word) for word in no_stop[i]]))
print(no_stem)
```

```python
#CountVectorizer using inBuilt
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

cv = CountVectorizer()
X = cv.fit_transform(no_stem)
CV = pd.DataFrame(data=X.toarray(),columns=cv.get_feature_names(),index=filename)
print(CV)
```

```python
#Tfidf Vectorizer using inBuilt
from sklearn.feature_extraction.text import TfidfVectorizer

cv = TfidfVectorizer()
X = cv.fit_transform(no_stem)
CV = pd.DataFrame(data=X.toarray(),columns=cv.get_feature_names(),index=filename)
print(CV)
```

```python
#Count Vectorizer without inBuilt
d = {}
unique_words = set()
for i in range(len(no_stem)):
  w = {}
  for word in no_stem[i].split():
    if word not in unique_words:
      unique_words.add(word)
    if word not in w:
      w[word] = 0
    w[word] += 1
  d[filename[i]] = w

CV = pd.DataFrame(d)
CV.fillna(0,inplace=True)
print(CV)
```

```python
#Tfidf without in built
import math
def idf(corpus,unique_word):
  idf_d = {}
  n = len(corpus)

  for word in unique_word:
    count = 0
    for sen in corpus:
      if word in sen.split():
        count += 1
    idf_d[word] = (math.log((1+n)/(count+1)))
  return idf_d

def compute_Tfidf(tf,idf):
  for file in tf:
    for word in tf[file]:
      tf[file][word] = (tf[file][word]*idf[word])/len(tf[file])
  return tf

TF = d
IDF = idf(no_stem,unique_words)
tf_idf = compute_Tfidf(TF,IDF)

TF_IDF = pd.DataFrame(tf_idf)
TF_IDF.fillna(0,inplace=True)
print(TF_IDF)
```

```python
#Corpus using Beautiful Soup
import requests
from bs4 import BeautifulSoup
filename_bs = []
corpus_bs = []
for i in range(10):
  a = 93985-i
  URL = "https://codeforces.com/blog/entry/"+str(a)
  filename_bs.append("blog/entry/"+str(a))
  r = requests.get(URL) 
  soup = BeautifulSoup(r.content, 'html5lib')
  for script in soup(["script", "style"]):
    script.extract()
  corpus_bs.append(soup.get_text())
print(corpus_bs)
```

```python
#Pre-processing

#Normalization
print("Normalization")
lower_bs = []
for i in range(len(corpus_bs)):
  lower_bs.append(" ".join(word.lower() for word in corpus_bs[i].split()))
print(lower_bs)
alpha_bs = []
for i in range(len(lower_bs)):
  alpha_bs.append(" ".join(word for word in lower_bs[i].split() if word.isalpha()))
print(alpha_bs)

#Tokenization
print("Tokenization")
from nltk import word_tokenize
tokenized_bs = []
for i in range(len(alpha_bs)):
  tokenized_bs.append(word_tokenize(alpha_bs[i]))
print(tokenized_bs)

#StopWord removal
print("StopWord removal")
from nltk.corpus import stopwords
stop = stopwords.words(fileids='english')
no_stop_bs = []
for i in range(len(tokenized_bs)):
  no_stop_bs.append([word for word in tokenized_bs[i] if word not in stop])
print(no_stop_bs)

#Stemming
print("Stemming")
from nltk.stem import PorterStemmer
no_stem_bs = []
ps = PorterStemmer()
for i in range(len(no_stop_bs)):
  no_stem_bs.append(' '.join([ps.stem(word) for word in no_stop_bs[i]]))
print(no_stem_bs)
```

```python
#Feature vector

cv_bs = CountVectorizer()
X_bs = cv_bs.fit_transform(no_stem_bs)
CV_bs = pd.DataFrame(data=X_bs.toarray(),columns=cv_bs.get_feature_names(),index=filename_bs)
print(CV_bs)
```
